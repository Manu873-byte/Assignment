```python
The ArithmeticError class in Python is a base class for errors that occur during arithmetic operations. It includes:

ZeroDivisionError – Raised when a number is divided by zero.
OverflowError – Raised when a calculation exceeds the maximum limit for a numeric type.
FloatingPointError – Occurs due to floating-point calculations (rare in Python).
Examples:
1. ZeroDivisionError:
a = 10 / 0 
2. OverflowError:
import math
print(math.exp(1000)) 
```

```python
The LookupError class is a base class for errors raised when an invalid key or index is used in a lookup operation. It includes:

KeyError – Raised when trying to access a dictionary key that doesn't exist.
IndexError – Raised when accessing an index that is out of range in a list or tuple.
Examples:
️1. KeyError:
data = {"name": "Alice", "age": 25}
print(data["gender"])
2. IndexError:
numbers = [10, 20, 30]
print(numbers[5])
```
